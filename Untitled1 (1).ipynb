# مقدمه‌ای بر تجزيه و تحليل سيستم‌ها

## مقدمه

در فصول قبلی توانایی فوق‌العاده تبدیل فوریه در تجزیه و تحلیل سیستم‌های خطی را مشاهده کردیم. این توانایی از آنجا ناشی می‌شود که بسیاری از سیگنال‌ها را می‌توان به صورت ترکیب خطی از توابع نمایی موهومی نوشت و چون توابع وی‍ژه سیستم‌ها نیز به صورت نمایی موهومی بودند، به‌سادگی توانستیم پاسخ به هر ورودی را به دست آوریم.

اما برخی سیگنال‌ها را نمی‌توان بر حسب توابع نمایی موهومی نوشت. در این فصل با معرفی **تبدیل لاپلاس**، که حالت کلی‌تر تبدیل فوریه است، می‌خواهیم بسط جدید سیگنال دلخواه را بر حسب توابع ویژه سیستم‌های LTI (توابع نمایی مختلط) ارائه کنیم.

## 6-1 تبدیل لاپلاس

برای یک سیستم LTI با پاسخ ضربه $h(t)$، دیدیم که پاسخ $y(t)$ سیستم به ورودی نمایی مختلط $e^{st}$ چنین می‌شود:

$$
y(t) = \int_{-\infty}^{+\infty} e^{s(t - \tau)} h(\tau)\, d\tau = e^{st} \int_{-\infty}^{+\infty} e^{-s\tau} h(\tau)\, d\tau = H(s) e^{st}
$$

که در آن:

$$
H(s) = \int_{-\infty}^{+\infty} e^{-s\tau} h(\tau)\, d\tau \tag{6-2}
$$

اگر $s = j\omega$ موهومی محض باشد، انتگرال معادله (6-2) همان **تبدیل فوریه** $h(t)$ بود. اما در حالت عمومی که $s$ عدد مختلط است، انتگرال (6-2) همان **تبدیل لاپلاس** $h(t)$ نامیده می‌شود که مانند تبدیل فوریه نقش مهمی در تجزیه و تحلیل سیگنال‌ها و سیستم‌های LTI دارد.

در حالت کلی، تبدیل لاپلاس سیگنال دلخواه $x(t)$ چنین تعریف می‌شود:

$$
X(s) \triangleq \int_{-\infty}^{+\infty} x(t) e^{-st}\, dt \tag{6-3}
$$

که در آن $s = \sigma + j\omega$ به‌طور کلی یک عدد مختلط است. انتگرال (6-3) همان‌طور که مشاهده می‌شود تابعی از $s$ است. گاهی اپراتور $\mathcal{L}$ را در مورد تبدیل لاپلاس به‌کار می‌برند و این‌گونه نمایش می‌دهند:

$$
x(t) \xleftrightarrow{\mathcal{L}} X(s)
$$

هنگامی که $s = j\omega$ باشد، معادله (6-3) به رابطه زیر تبدیل می‌شود:

$$
X(j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-j\omega t}\, dt \tag{6-4}
$$

این همان **تبدیل فوریه** $x(t)$ است که به صورت $X(\omega)$ قبلاً نشان داده شد. بنابراین تبدیل لاپلاس وقتی $s = j\omega$ باشد، مساوی تبدیل فوریه $x(t)$ خواهد بود:

$$
|X(s)|_{s = j\omega} = \mathcal{F}\{x(t)\} = X(\omega) \tag{6-5}
$$

ارتباط جالب دیگری میان تبدیل لاپلاس و تبدیل فوریه وجود دارد که اگر معادله (6-3) را به صورت معادله (6-6) بنویسیم، می‌توان این ارتباط را پیدا کرد:

$$
X(\sigma + j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-(\sigma + j\omega)t}\, dt \tag{6-6}
$$


# فصل ششم: تبدیل لاپلاس

یا  
$$
X(\sigma + j\omega) = \int_{-\infty}^{+\infty} x(t) e^{-(\sigma + j\omega)t} \, dt \tag{6-7}
$$

عبارت فوق، $X(s)$ را برحسب نقطه‌ای از صفحه مختلط تعریف می‌کند که $s = \sigma + j\omega$ است.  
از آن‌جا که تبدیل لاپلاس همان تبدیل فوریه برای $x(t)e^{-\sigma t}$ است، بسیاری از نتایج سیگنال‌های با مدت زمان محدود را می‌توان از تحلیل فوریه برای $x(t)e^{-\sigma t}$ به‌دست آورد.

---

## مثال

فرض کنید:

$$
x(t) = e^{-at} u(t), \quad a > 0 \tag{6-8}
$$

در این‌صورت تبدیل لاپلاس آن از رابطه‌ی (6-7) چنین محاسبه می‌شود:

$$
X(\omega) = \int_{-\infty}^{+\infty} e^{-at} e^{-j\omega t} \, dt = \frac{1}{j\omega + a}, \quad a > 0 \tag{6-9}
$$

و از معادله (6-7) داریم:

$$
X(s) = \int_{0}^{\infty} e^{-at} e^{-st} \, dt = \int_{0}^{\infty} e^{-(s+a)t} \, dt \tag{6-10}
$$

با فرض $s = \sigma + j\omega$ داریم:

$$
X(\sigma + j\omega) = \int_{0}^{\infty} e^{-(\sigma + a)t} e^{-j\omega t} \, dt \tag{6-11}
$$

می‌بینیم که تابع $x(t)$ همان است و از معادله (6-9) می‌توان نتیجه گرفت که:

$$
X(s) = \frac{1}{\sigma + a + j\omega}, \quad (\sigma + a) > 0 \tag{6-12}
$$

یا بطور معادل:

$$
X(s) = \frac{1}{s + a}, \quad \text{Re}(s) > -a \tag{6-13}
$$

---

## بررسی تابع دیگر:

اگر:

$$
x(t) = e^{at} u(t), \quad a > 0
$$

تبدیل فوریه آن وجود ندارد چون $x(t)$ محدود نیست.  
اما تبدیل لاپلاس آن برای $\text{Re}(s) > a$ برابر است با:

$$
X(s) = \frac{1}{s - a} \tag{6-14}
$$

---

در این حالت، $X(s)$ فقط برای بخش حقیقی بزرگ‌تر از $a$ تعریف شده است.

اگر $x(t) = e^{at}$ باشد و فقط برای $t > 0$ تعریف شود، باز هم تبدیل لاپلاس وجود ندارد زیرا ناحیه همگرایی شامل $\text{Re}(s) = 0$ نخواهد شد.


### مثال (6-3): مقایسه با مثال قبلی انجام دهید

اگر:
$$
x(t) = e^{-at} u(-t) \tag{6-15}
$$

بنویسید:
$$
X(s) = \int_{-\infty}^{0} e^{-at} e^{-st} \, dt = \int_{-\infty}^{0} e^{-(s + a)t} \, dt \tag{6-16}
$$

و با:
$$
X(s) = \frac{1}{s + a}, \quad \text{Re}\{s\} < -a \tag{6-17}
$$

توجه کنید که برای همگرا شدن تبدیل لاپلاس باید داشته باشیم:
$$
\text{Re}\{s\} < -a \tag{6-18}
$$

بنابراین:
$$
\mathcal{L}\{e^{-at} u(-t)\} = \frac{1}{s + a}, \quad \text{Re}\{s\} < -a \tag{6-19}
$$

عبارت تبدیل لاپلاس در هر دو مثال (6-3) و (6-1) یکی است اما ناحیه‌ی همگرایی (ROC) آن‌ها متفاوت است. این موضوع نشان می‌دهد که عبارت تبدیل لاپلاس به تنهایی برای بازسازی تابع زمانی کافی نیست، بلکه ناحیه همگرایی نیز باید مشخص شود.

---

در حالت کلی، بسیاری از سیگنال‌های پرکاربرد، دارای تبدیل لاپلاس یکسان اما نواحی همگرایی متفاوتی هستند. بنابراین تحلیل ناحیه همگرایی ROC بسیار اهمیت دارد.

اگر سیگنال $x(t)$ شامل هر دو قسمت $u(t)$ و $u(-t)$ باشد (مثلاً تابع زوج یا متناوب)، آنگاه تبدیل لاپلاس آن تنها زمانی موجود است که نواحی همگرایی دو قسمت با هم اشتراک داشته باشند.

مثلاً سیگنال ترکیبی از دو تابع مثال‌های (6-1) و (6-3) یعنی:
$$
x(t) = e^{-at}, \quad \text{برای تمام t}
$$

تبدیل لاپلاس آن وجود ندارد چون ROC مربوط به قسمت راست $\text{Re}\{s\} > -a$ و برای قسمت چپ $\text{Re}\{s\} < -a$ است، و هیچ اشتراکی ندارند.

---

**شکل (6-3)** روش نمایش ناحیه همگرایی (ROC) را برای دو حالت نشان می‌دهد:

- a) مربوط به مثال (6-1)
- b) مربوط به مثال (6-3)

> **نکته:** ROC = Region of Convergence


```python
import numpy as np
import matplotlib.pyplot as plt

# مقدار a
a = 2

# تنظیم شکل
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# ================================
# حالت a) Re{s} > -a (مثال 6-1)
axs[0].axvline(-a, color='black', linestyle='--')
axs[0].fill_betweenx([-10, 10], -a, 5, color='gray', alpha=0.4)
axs[0].set_xlim(-5, 5)
axs[0].set_ylim(-10, 10)
axs[0].set_title("a) Re{s} > -a")
axs[0].set_xlabel(r"$\sigma$")
axs[0].set_ylabel(r"$j\omega$")
axs[0].grid(True)
axs[0].text(-a + 0.1, 9, r"$\sigma = -a$", rotation=90)

# ================================
# حالت b) Re{s} < -a (مثال 6-3)
axs[1].axvline(-a, color='black', linestyle='--')
axs[1].fill_betweenx([-10, 10], -5, -a, color='gray', alpha=0.4)
axs[1].set_xlim(-5, 5)
axs[1].set_ylim(-10, 10)
axs[1].set_title("b) Re{s} < -a")
axs[1].set_xlabel(r"$\sigma$")
axs[1].set_ylabel(r"$j\omega$")
axs[1].grid(True)
axs[1].text(-a + 0.1, 9, r"$\sigma = -a$", rotation=90)

plt.tight_layout()
plt.show()
```

```python
## فصل ششم: تبدیل لاپلاس

مثال 1: یک سیگنال متناوب مانند در شکل (1-6)، ممکن است متشکل باشد از چندین u(t) با شیفت داده شده.  
فرض کنید:

$$
x(t) = e^{-t}u(t) + e^{-2t}u(t)
$$

عبارت تبدیل لاپلاس چنین است:

$$
X(s) = \int_{-\infty}^{\infty} \left[e^{-t}u(t) + e^{-2t}u(t)\right] e^{-st} dt
$$

$$
= \int_{0}^{\infty} e^{-(s+1)t} dt + \int_{0}^{\infty} e^{-(s+2)t} dt \tag{6-19}
$$

اگر در هر دو عبارت \( \text{Re}(s) > -1 \) و \( \text{Re}(s) > -2 \) در نظر گرفته شود، نتیجه نهایی برابر می‌شود با:

$$
X(s) = \frac{1}{s+1} + \frac{1}{s+2} \tag{6-20}
$$

برای تعیین ROC (مکان ناحیه همگرایی)، باید مجموع دو تابع خطی را در نظر بگیریم. در حالی که هر کدام از تبدیل‌های لاپلاس دارای یک ناحیه خاصی از s هستند، مجموع نواحی اشتراکی باید بررسی شود.  
برای این تابع داریم:

$$
\mathcal{L}\{x(t)\} = \frac{1}{s+1} + \frac{1}{s+2}, \quad \text{ROC: } \text{Re}(s) > -1 \tag{6-21}
$$

مثال دیگر:

$$
x(t) = e^{2t}u(-t)
$$

عبارت تبدیل لاپلاس آن:

$$
X(s) = \int_{-\infty}^{0} e^{2t} e^{-st} dt = \int_{-\infty}^{0} e^{-(s - 2)t} dt
$$

در این صورت:

$$
X(s) = \frac{1}{s - 2}, \quad \text{Re}(s) < 2 \tag{6-22}
$$

نکته:  
یک تابع مختلط کسری را می‌توان به صورت:

$$
X(s) = \frac{2s + 3}{s^2 + 2s + 1} = \frac{N(s)}{D(s)} \tag{6-23}
$$

کسی است که در آن \( N(s) \) و \( D(s) \) توابع چندجمله‌ای حقیقی از \( s \) هستند. در فصل بعدی نشان خواهیم داد که اگر \( X(s) \) ترکیب توابع کسری با صورت و مخرج چندجمله‌ای باشد، می‌توان آن را به عنوان پاسخ ضربه‌ای یک سیستم LTI که با معادلات دیفرانسیل خطی تعریف شده، در نظر گرفت.

---

<sup>1</sup> S-Plane  
<sup>2</sup> Rational
```

```python
## مقدمه‌ای دیگر بر تحلیل سیستم‌ها

تبدیل \( H(s) \) مربوط به کسری شامل صورت‌های \( N(s) \) (صورت‌ها) و مخرج‌های \( D(s) \) (مخرج‌ها) تبدیل هستند و محدوده‌ای در ROC آن ناحیه مشخصی می‌تواند برای مثال در شکل (6-6) رسم شود.

### ROC مربوط به شکل (6-6):

**شکل 6-6**: نمودار صفر و قطب، و ROC برای یک تابع \( X(s) \)

علامت × جهت نمایش قطب و علامت ○ جهت نمایش صفر بکار گرفته شده‌اند.

مثال 2: فرض کنید:

$$
x(t) = -\frac{4}{3}e^{-t}u(t) + \frac{1}{3}e^{-2t}u(t) \tag{6-7}
$$

تبدیل لاپلاس \( x(t) \) برای سیگنال بالا طبق تعریف تبدیل لاپلاس به صورت زیر خواهد بود:

$$
X(s) = \frac{4}{s+1} + \frac{1}{s+2}, \quad \text{Re}(s) > 2 \tag{6-8}
$$

ناحیه همگرایی از نظر عملی تک تک توان‌هایی سمت راست محور مختلط بوده که همراهی تبدیل برای زمانی‌های مثبت در نظر گرفته شده است. توجه کنید که همگرایی (معادله 6-8) وابسته به تابع \( x(t) \) می‌باشد.

$$
X(s) = \frac{(s - 1)^2}{(s+1)(s-2)} \quad \text{Re}(s) > 2 \tag{6-9}
$$

### رسم صفر و قطب، و ROC برای \( X(s) \) بصورت شکل (6-6) می‌باشد.

مطابق آنچه که در بالا نشان داده شد، هر تبدیل شامل صفرها و قطب‌هایی است. همگرایی در ROC شامل تمام نقاط s در صفحه مختلط است که موجب همگرایی انتگرال تبدیل لاپلاس می‌گردد.  
وجود ناحیه‌ای که در آن تبدیل لاپلاس وجود داشته باشد مستلزم این است که تابع نمایی زمان پیوسته از نوع نمایی نزولی باشد.

---

<sup>1</sup> Zeros  
<sup>2</sup> Poles
```

```python
import matplotlib.pyplot as plt
import numpy as np

# مقدار قطب‌ها و صفرها
poles = [-1, -2]
zeros = []

# ناحیه ROC (Re(s) > -2)
roc_left = -2
roc_right = -1

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 6))

# رسم قطب‌ها (×)
for p in poles:
    plt.plot(np.real(p), np.imag(p), 'rx', markersize=12, label='Pole' if p == poles[0] else "")

# رسم صفرها (○)
for z in zeros:
    plt.plot(np.real(z), np.imag(z), 'go', markersize=10, label='Zero' if z == zeros[0] else "")

# رسم ناحیه ROC با رنگ خاکستری
plt.axvspan(roc_left, roc_right, color='gray', alpha=0.3, label='ROC')

# اضافه کردن محورهای x و y
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# برچسب‌ها و عنوان
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.title('Pole-Zero Plot with ROC')

# محدودیت‌های نمودار
plt.xlim(-4, 2)
plt.ylim(-3, 3)

# گرید و legend
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

```python
### فصل ششم: تبدیل لاپلاس

در این مثال دو قطب \( s = -1 \) و \( s = -2 \) در صفحه s قرار دارند و ناحیه همگرایی (ROC) سمت راست بزرگترین قطب قرار گرفته است.

گفته شد برای بسیاری از تبدیل‌های لاپلاس معمولی (مثل پاسخ سیستم‌ها)، ناحیه ROC وجود دارد و شکل آن می‌تواند هموار باشد تا بتوان به راحتی مشخصه قطب‌ها، ناحیه ROC، و ساختار تابع تبدیل را تحلیل کرد (مانند شکل زیر).

---

### ۶-۴ خواص ناحیه همگرایی برای تبدیل لاپلاس

تبدیل لاپلاس \( X(s) \) فقط وقتی وجود دارد که تبدیل انتگرالی در تعریف آن دارای حد متناهی باشد. این اتفاق زمانی می‌افتد که تابع \( x(t) \) با سرعت کافی به سمت صفر میل کند و همزمان \( s \) به گونه‌ای انتخاب شود که انتگرال:

\[
X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt
\]

همگرا باشد. شکل زیر چند حالت مختلف از ناحیه ROC برای سیگنال‌هایی با خواص متفاوت را نشان می‌دهد.
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# تنظیمات مشترک
poles = [[-1, -2], [-2, -3], [-1, -3]]
rocs = [[-1, 3], [-3, -2], [-3, -1]]  # [left, right] for each plot
titles = ["ROC سمت راست", "ROC سمت چپ", "ROC بین دو قطب"]

for i in range(3):
    ax = axs[i]
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal', adjustable='box')
    ax.set_title(titles[i])
    ax.set_xlabel("Re")
    ax.set_ylabel("Im")

    # قطب‌ها
    for p in poles[i]:
        ax.plot(np.real(p), 0, 'rx', markersize=12)

    # ROC
    ax.axvspan(rocs[i][0], rocs[i][1], color='gray', alpha=0.4)

plt.tight_layout()
plt.show()
```
